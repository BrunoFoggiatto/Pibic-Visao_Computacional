```python
import os
import shutil
import random

orig_train_dir = "/content/drive/MyDrive/Colab Notebooks/Dataset/train"
orig_val_dir = "/content/drive/MyDrive/Colab Notebooks/Dataset/val"
orig_test_dir = "/content/drive/MyDrive/Colab Notebooks/Dataset/test"

new_dataset_dir = "/content/drive/MyDrive/Colab Notebooks/Dataset_250"
new_train_dir = os.path.join(new_dataset_dir, "train")
new_val_dir = os.path.join(new_dataset_dir, "val")

for dir in [new_train_dir, new_val_dir]:
    os.makedirs(dir, exist_ok=True)

def copy_fixed_number(source_dir, target_dir, num_files):
    files = [f for f in os.listdir(source_dir) if f.endswith(".jpg") or f.endswith(".png")]
    files = sorted(files)
    num_files = min(num_files, len(files))
    sampled_files = random.sample(files, num_files)

    for file in sampled_files:
        base_name, ext = os.path.splitext(file)
        image_path = os.path.join(source_dir, file)
        text_path = os.path.join(source_dir, base_name + ".txt")

        shutil.copy(image_path, os.path.join(target_dir, file))
        if os.path.exists(text_path):
            shutil.copy(text_path, os.path.join(target_dir, base_name + ".txt"))

copy_fixed_number(orig_train_dir, new_train_dir, 400)
copy_fixed_number(orig_val_dir, new_val_dir, 400)

train_size = len([f for f in os.listdir(new_train_dir) if f.endswith(".jpg") or f.endswith(".png")])
val_size = len([f for f in os.listdir(new_val_dir) if f.endswith(".jpg") or f.endswith(".png")])
```

```python
from google.colab import drive
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Montar o Google Drive
drive.mount('/content/drive')

# Caminhos
model_path = "/content/drive/MyDrive/Colab Notebooks/Yolo/best_trained.pt"
val_images_path = "/content/drive/MyDrive/Colab Notebooks/dataset_small/test"
val_labels_path = "/content/drive/MyDrive/Colab Notebooks/dataset_small/test"
save_path = "/content/drive/MyDrive/Colab Notebooks/Yolo/resultados_fp_fn.csv"

# Carregar modelo
model = YOLO(model_path)

# Função IoU
def calculate_iou(box1, box2):
    x1, y1, x2, y2 = box1
    x1g, y1g, x2g, y2g = box2
    xi1, yi1 = max(x1, x1g), max(y1, y1g)
    xi2, yi2 = min(x2, x2g), min(y2, y2g)
    inter_area = max(0, xi2 - xi1) * max(0, yi2 - yi1)
    box1_area = (x2 - x1) * (y2 - y1)
    box2_area = (x2g - x1g) * (y2g - y1g)
    union = box1_area + box2_area - inter_area
    return inter_area / union if union > 0 else 0

# Exibição
def display_image_with_boxes(image_path, filename, boxes, fp_count, fn_count):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    for box, color, label in boxes:
        x1, y1, x2, y2 = box
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
        cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
    plt.figure(figsize=(8, 6))
    plt.imshow(image)
    plt.title(f"{filename} | FP: {fp_count} | FN: {fn_count}", fontsize=12)
    plt.axis("off")
    plt.show()

# Função principal
def check_fp_fn(val_images_path, val_labels_path):
    images = [f for f in os.listdir(val_images_path) if f.endswith((".jpg", ".png"))]
    images = images[:60]

    resultados = []

    for filename in images:
        image_path = os.path.join(val_images_path, filename)
        label_path = os.path.join(val_labels_path, filename.replace(".jpg", ".txt").replace(".png", ".txt"))

        # Predição
        results = model(image_path)
        predictions = results[0].boxes.xyxy.cpu().numpy()

        # Ground truth
        ground_truths = []
        if os.path.exists(label_path):
            with open(label_path, "r") as f:
                for line in f:
                    parts = list(map(float, line.split()))
                    if len(parts) == 5:
                        _, x, y, w, h = parts
                        H, W, _ = cv2.imread(image_path).shape
                        x1 = (x - w / 2) * W
                        y1 = (y - h / 2) * H
                        x2 = (x + w / 2) * W
                        y2 = (y + h / 2) * H
                        ground_truths.append((x1, y1, x2, y2))

        detected = [False] * len(ground_truths)
        boxes = []
        fp, fn = 0, 0

        for pred in predictions:
            x1p, y1p, x2p, y2p = pred[:4]
            iou_max, best_match = 0, -1
            for i, gt in enumerate(ground_truths):
                iou = calculate_iou([x1p, y1p, x2p, y2p], gt)
                if iou > iou_max:
                    iou_max = iou
                    best_match = i
            if iou_max < 0.5:
                fp += 1
                boxes.append(([x1p, y1p, x2p, y2p], (255, 0, 0), "FP"))
            else:
                detected[best_match] = True
                boxes.append(([x1p, y1p, x2p, y2p], (0, 255, 0), "OK"))

        for i, gt in enumerate(ground_truths):
            if not detected[i]:
                fn += 1
                boxes.append((gt, (0, 0, 255), "FN"))

        tp = sum(detected)

        resultados.append({
            "imagem": filename,
            "fp": fp,
            "fn": fn,
            "tp": tp,
            "total_gt": len(ground_truths),
            "total_pred": len(predictions)
        })

        if fp > 0 or fn > 0:
            display_image_with_boxes(image_path, filename, boxes, fp, fn)

    # Salvar CSV (Visualização do result de cada imagem)
    df = pd.DataFrame(resultados)
    df.to_csv(save_path, index=False)

    # Métricas totais
    total_tp = df["tp"].sum()
    total_fp = df["fp"].sum()
    total_fn = df["fn"].sum()

    precision = total_tp / (total_tp + total_fp) if (total_tp + total_fp) > 0 else 0
    recall = total_tp / (total_tp + total_fn) if (total_tp + total_fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    print("\n RESULTADOS: \n")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")

# Executar
check_fp_fn(val_images_path, val_labels_path)
```
